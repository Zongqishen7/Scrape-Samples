```python
def get_news_today(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    divs = soup.find_all("div", {"class": "DBQmFf NclIid BL5WZb Oc0wGc xP6mwf j7vNaf"})
    companys = []
    newsheads = []
    h_4s = []
    # Loop news
    for i in divs:
        company = i.find("a", {"class": "wEwyrc AVN2gc uQIVzc Sksgp"}).get_text()
        newshead = i.find("h3").get_text()
        h_4 = i.find("h4").get_text()
        companys.append(company)
        newsheads.append(newshead)
        h_4s.append(h_4)

    dict = {
        "companys": companys,
        "newsheads": newsheads,
        "h_4s": h_4s
    }

    news_today = pd.DataFrame(dict)
    return news_today


url = 'https://news.google.com/topics/CAAqIggKIhxDQkFTRHdvSkwyMHZNRGxqTjNjd0VnSmxiaWdBUAE?hl=en-US&gl=US&ceid=US%3Aen'
news = get_news_today(url)
news
```
