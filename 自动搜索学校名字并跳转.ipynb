# 简易地获取网站并点击


```python
def click_url(search_item):
    driver.get('https://www.google.com')
    time.sleep(0.2)
    m = driver.find_element_by_xpath('//input[@class="gLFyf gsfi"]')
    m.send_keys(str(search_item))
    time.sleep(0.1)
    m.send_keys(Keys.ENTER)
```
