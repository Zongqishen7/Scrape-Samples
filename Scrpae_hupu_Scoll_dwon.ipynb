```python

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.hupu.com/')


import time
# scroll down 5 times
for i in range(7):
    driver.execute_script("window.scrollTo(0, \
    document.body.scrollHeight);")

    # wait for the content to be loaded
    time.sleep(2)

text = driver. \
    find_elements_by_css_selector("div.item-title-conent")

news = []
for i, q in enumerate(text):
    print(i, q.text)
    print("\n")
    news.append(q.text)
```
